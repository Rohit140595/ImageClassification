```python
from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# loading the cifar10 dataset
from keras.datasets import cifar10

# load dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# summarize loaded dataset
print('Train: X=%s, y=%s' % (X_train.shape,y_train.shape))
print('Test: X=%s, y=%s' % (X_test.shape, y_test.shape))
```

## Here we are showing before and after conversion of 3 image samples only

```python
X_train = X_train[:3]
```

**This is what we have used for our project. Keeping the features generated from HOG into feasible limit for our computing systems.**

**Below configuration will generate 3780 features per image.**

```python
for i in range(len(X_train)):
    plt.subplot(1,2,1)
    img = X_train[i]
    plt.imshow(img)
    resized_img = resize(img, (128, 64))
    fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2),
                        visualize=True, multichannel=True, transform_sqrt=True, block_norm="L1")
    plt.subplot(1,2,2)
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
    plt.imshow(hog_image_rescaled, cmap=plt.cm.gray)
    
    plt.show()
```

## **Future expansion:**  
**Depending on the availability of the high speed computing power, one of the following iterations can be applied for extracting more powerful HOG features that will definitely increase ML algorithm's accuracy.**

**For low end systems:**

```python
for i in range(len(X_train)):
    plt.subplot(1,2,1)
    img = X_train[i]
    plt.imshow(img)
    
    resized_img = resize(img, (64,64))
    fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2),
                        visualize=True, multichannel=True, transform_sqrt=True, block_norm="L1")
    plt.subplot(1,2,2)
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
    plt.imshow(hog_image_rescaled, cmap=plt.cm.gray)
    
#     plt.savefig("hog_images{}.png".format(i), dpi = 300)
    plt.show()
```

**For powerful systems:**

```python
for i in range(len(X_train)):
    plt.subplot(1,2,1)
    img = X_train[i]
    plt.imshow(img)
    
    resized_img = resize(img, (64,64))
    fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(4, 4), cells_per_block=(2, 2),
                        visualize=True, multichannel=True, transform_sqrt=True, block_norm="L1")
    plt.subplot(1,2,2)
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
    plt.imshow(hog_image_rescaled, cmap=plt.cm.gray)
    
    plt.show()
```

**For extremely powerful systems:**

**Below configuration will generate 34,596 features per image.**

```python
for i in range(len(X_train)):
    plt.figure(figsize = (8, 8))
    plt.subplot(1,2,1)
    img = X_train[i]
    plt.imshow(img)
    
    resized_img = resize(img, (128, 128))
    fd, hog_image = hog(resized_img, orientations=36, pixels_per_cell=(4, 4), cells_per_block=(2, 2),
                        visualize=True, multichannel=True, transform_sqrt=True, block_norm="L1")    
    plt.subplot(1,2,2)
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
    plt.imshow(hog_image_rescaled, cmap=plt.cm.gray)
    
#     plt.savefig("hog_images_expansion{}.png".format(i), dpi = 300)    
    plt.show()
```
